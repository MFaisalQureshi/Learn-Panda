```python
import pandas as pd
pakistan_weather=pd.DataFrame({
    'city':['Karachi','Lahore','Islamabad'],
    'tempearture':[35,30,20],
    'event':['sunny','sunny','cloudy']
})
pakistan_weather
```

```python
india_weather=pd.DataFrame({
    'city':['mumbai','banglore','delhi'],
    'tempearture':[40,35,44],
    'event':['sunny','cloudy','sunny']
})
india_weather
```

```python
df=pd.concat([pakistan_weather,india_weather],ignore_index=True)
df
```

```python
df=pd.concat([pakistan_weather,india_weather],keys=['Pakistan','India'])
df
```

```python
temperature_df=pd.DataFrame({
    'city':['Karachi','Lahore','Islamabad'],
    'tempearture':[35,30,20]
})
temperature_df
```

```python
windspead_df=pd.DataFrame({
    'city':['Islamabad','Karachi','Lahore'],
    'windspead':[10,20,30]
},index=[2,0,1])
windspead_df
```

```python
df=pd.concat([temperature_df,windspead_df],axis=1)
df
```
