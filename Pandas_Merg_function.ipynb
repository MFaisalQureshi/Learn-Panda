```python
import pandas as pd
df1=pd.DataFrame({
  'city':['Karachi','Lahore','Islamabad','Multan','Faisalabad'],
  'temperature':[30,35,40,35,38]  

})
df1
```

```python
df2=pd.DataFrame({
    'city':['Karachi','Lahore','Islamabad'],
    'humidity':[10,20,30]
})
df2
```

```python
#Merge Basically Perform Intersection Between Two Data sets
df=pd.merge(df1,df2,on='city')
df
```

```python
df=pd.merge(df1,df2,on='city', how="outer")
df
```
